```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from statistics import mean

warnings.filterwarnings("ignore")

```

```python

dataset = pd.read_excel(r"D:\GitHub\House_Rent_Prices_with_Scraping\Kira_Ev_Dataset_modified.xlsx")

dupl_entr_1 = dataset[dataset.duplicated()] # find duplicate entries
print("Dupicate Entries: ", dupl_entr_1.shape)
dataset = dataset.drop_duplicates() # delete duplicate entries

dataset
```

```python
mean_rent = mean(dataset["Fiyat"])
mean_rent = int(round(mean_rent,0))

print("Mean of House Rental Prices in Turkey:", mean_rent,"TL \n")

print("Information about the Cheapest Rental House:\n\n",dataset.loc[dataset['Fiyat'].idxmin()], "\n")

print("Information about the Most Expensive Rental House:\n\n",dataset.loc[dataset['Fiyat'].idxmax()], "\n")
```

```python
plt.subplots(figsize=(25,12))
sns.barplot(data=dataset, y="Fiyat", x="Konum", palette="rocket")
plt.xticks(rotation=90)
plt.title('Terrorist Attacks Each Target')
plt.show()
```
